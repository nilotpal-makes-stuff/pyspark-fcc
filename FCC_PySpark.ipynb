## 1. Introduction to PySpark

PySpark is the Python API for Apache Spark, designed to enable large-scale distributed data processing and analytics using Python. It includes **Spark SQL** for querying databases, **DataFrames** for structured storage of data, **Structured Streaming** for processing continuous data streams as unbounded tables, **MLlib** for machine learning and **GraphX** for graph processing.

Key features of PySpark :
* **Distributed Computing**: PySpark runs computations in parallel across a cluster, enabling the processing of petabyte-scale datasets that cannot fit on a single machine.
* **Fault Tolerance & Optimization** : It utilizes **Resilient Distributed Datasets (RDDs)** for automatic data recovery and employs lazy evaluation to optimize execution plans before actions are triggered.
* **Versatility**: It supports both batch and real-time processing, integrates seamlessly with the Python ecosystem (e.g., pandas, NumPy), and runs on various cluster managers like YARN, Kubernetes, and Mesos.

By default when reading data from a file to a dataframe, Spark will set all columns datatype as 'string'. We can use the **inferSchema=True** in **csv()** method to make Spark infer the schema of each column of the dataframe.

```python
#Create dataframe from csv file
#read contains reader functions for different types of dataset files
diabetes_df = spark.read.csv("diabetes.csv", header=True, inferSchema=True)
print(type(diabetes_df))
diabetes_df
```

We can also use the **"DataFrame.describe()"** method to get a summary on the data inside the Dataframe. This returns a dataframe object and can be viewed using **"Dataframe.show()"** method.

describe() generates the following statistical values for summarizing the data :
* count - count of values(null included)
* mean - average value of the column
* stddev - standard deviation
* min - minimum value
* max - maximum value

Columns having string datatype are also considered in the describe method, mean and stddev are null for string datatype columns as mean and stddev for string values have no meaning.

```python
print(diabetes_df.describe(), type(diabetes_df.describe()), sep='\n')
print()

#Display describe summary statistics
diabetes_desc_df = diabetes_df.describe()
diabetes_desc_df.show()
```

To specify the number of decimal places for floating point results, select the columns and use **format_number()** method from the **pyspark.sql.functions** package on the column and specify the number of decimal places. **format_number()** method takes in "column" and "number of decimal places" as parameters. format_number returns a string value and not a floating point value.

We need to cast the column to float datatype using **cast()** method on column object - get column object using **col()** method.

We can also use the **"alias()"** method after format_number() to rename the column.

```python
#Limit decimal places
from pyspark.sql.functions import col, format_number

formatted_desc_df = diabetes_desc_df.select(
col('summary'),
format_number(col('Pregnancies').cast('float'), 2).alias('Pregnancies'),
format_number(col('Glucose').cast('float'), 2).alias('Glucose'),
format_number(col('BloodPressure').cast('float'), 2).alias('BloodPressure'),
format_number(col('SkinThickness').cast('float'), 2).alias('SkinThickness'),
format_number(col('Insulin').cast('float'), 2).alias('Insulin'),
format_number(col('BMI').cast('float'), 2).alias('BMI'),
format_number(col('DiabetesPedigreeFunction').cast('float'), 2).alias('DiabetesPedigreeFunction'),
format_number(col('Age').cast('float'), 2).alias('Age'),
format_number(col('Outcome').cast('float'), 2).alias('Outcome'),
)

formatted_desc_df.show()
```

We can also use a much simpler method **round()** from **pyspark.sql.functions**. round() takes in the column and the number of decimal places to round. Output data is a float value.

```python
#Limit decimal places using round from pyspark
from pyspark.sql.functions import round as pyspark_round, col

formatted_desc_df2 = diabetes_desc_df.select(
col('summary'),
pyspark_round(col('Pregnancies'), 2).alias('Pregnancies'),
pyspark_round(col('Glucose'), 2).alias('Glucose'),
pyspark_round(col('BloodPressure'), 2).alias('BloodPressure'),
pyspark_round(col('SkinThickness'), 2).alias('SkinThickness'),
pyspark_round(col('Insulin'), 2).alias('Insulin'),
pyspark_round(col('BMI'), 2).alias('BMI'),
pyspark_round(col('DiabetesPedigreeFunction'), 2).alias('DiabetesPedigreeFunction'),
pyspark_round(col('Age'), 2).alias('Age'),
pyspark_round(col('Outcome'), 2).alias('Outcome'),
)

formatted_desc_df2.show()
```

## 3. PySpark handling missing values

We can deal with null values using the following methods :
* drop
* fill
* replace

We can use **Dataframe.na.drop()** to drop all rows/columns with null values, the logic of dropping depends on the parameters provided.
* how - 'any', 'all'
    - 'any' drop rows with any null value in row
    - 'all' drop rows if all null values in row
* thresh - int value - drop row if minimum 'thresh' number of non null values are not present.
* subset - list of column names - columns to search for null values

Lets drop null values from a dataframe using **how='any'** and **how='all'**.

```python
# Wine dataframe
# This dataframe contains null values
wine_df = spark.read.csv("winemag-data-130k-v2.csv", header=True, inferSchema=True)
wine_df.show(5)
```

```python
print(f"Number of rows : {wine_df.count()}")
print()

#Dropping all rows with atleast 1 null value - using how='any'
wine_df_dropanynull = wine_df.na.drop(how='any')
wine_df_dropanynull.show(5)
print(f"Number of rows after dropping rows with atleast 1 null values : {wine_df_dropanynull.count()}")
print()

#Dropping all rows with all null values - using how='all'
wine_df_dropallnull = wine_df.na.drop(how='all')
wine_df_dropallnull.show(5)
print(f"Number of rows after dropping rows with all null values : {wine_df_dropallnull.count()}")
print()
```

We can also use **thresh = int_value** to determine the minimum number of non null values required for a row to not be dropped.

```python
# At least 10 columns need to have values
wine_df_dropthreshnull = wine_df.na.drop(thresh=10)
wine_df_dropthreshnull.show(5)
print(f"Number of rows after dropping rows not meeting the threshold : {wine_df_dropthreshnull.count()}")
```

We can also use **subset = list_of_columns** to apply null value search logic only on columns in the subset.

```python
wine_df_dropsubsetnull = wine_df.na.drop(subset=['price'])
wine_df_dropsubsetnull.show(5)
print(f"Number of rows after dropping rows containing null values in 'price' column : {wine_df_dropsubsetnull.count()}")
```
