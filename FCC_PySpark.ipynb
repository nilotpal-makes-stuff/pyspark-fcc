## 1. Introduction to PySpark

PySpark is the Python API for Apache Spark, designed to enable large-scale distributed data processing and analytics using Python. It includes **Spark SQL** for querying databases, **DataFrames** for structured storage of data, **Structured Streaming** for processing continuous data streams as unbounded tables, **MLlib** for machine learning and **GraphX** for graph processing.

Key features of PySpark :
* **Distributed Computing**: PySpark runs computations in parallel across a cluster, enabling the processing of petabyte-scale datasets that cannot fit on a single machine.
* **Fault Tolerance & Optimization** : It utilizes **Resilient Distributed Datasets (RDDs)** for automatic data recovery and employs lazy evaluation to optimize execution plans before actions are triggered.
* **Versatility**: It supports both batch and real-time processing, integrates seamlessly with the Python ecosystem (e.g., pandas, NumPy), and runs on various cluster managers like YARN, Kubernetes, and Mesos.

When working with Spark, we first need to initialise a Spark Session using the **SparkSession** object.

```python
from pyspark.sql import SparkSession
```

```python
#Creating spark session
spark = SparkSession.builder.appName("Practice").getOrCreate()
```

Here
* **appName()** method is used to specify name of Application
* **getOrCreate()** method either creates a Spark Session or retreives an already existing Spark Session.

We can display SparkSession details using the variable which contains the SparkSession, in this case **spark**. This displays the following details :
* Spark Context
* Spark Version
* Spark Master - contains information on nodes of spark session cluster
* AppName - Name of the application

```python
#Session Details
spark
#Version
#Master contains nodes of the PySpark session cluster
#AppName is name of the app provided above
```

Data in files can be read into PySpark DataFrames using the **SparkSession.read** object. This object contains multiple methods for each file type which contain data. To read csv file into pyspark dataframe use **SparkSession.read.csv()** method.

We can also use **.option()** method to specify certain options which apply onto the file reading operation. For example 'header','true' sets first row of csv file as header of the dataframe.

Alternatively, the **csv()** method contains keyword parameter **header** for setting the first row of the csv file as header of the dataframe.

```python
#Create dataframe from csv file
#read contains reader functions for different types of dataset files
diabetes_df = spark.read.csv("diabetes.csv", header=True)
print(type(diabetes_df))
diabetes_df
```

To display dataframe contents, use the **DataFrame.show(N)** method to display first N rows of the dataframe. By default the show method displays the first 20 rows of the dataframe.

```python
#Display dataframe - default shows first 20 rows
diabetes_df.show()
```

We can display the schema of the PySpark DataFrame using the **DataFrame.printSchema()** method. This displays the schema of the dataframe - datatype of each column and whether each column is nullable.

```python
diabetes_df.printSchema()
```

## 2. Working with Pyspark DataFrames

By default when reading data from a file to a dataframe, Spark will set all columns datatype as 'string'. We can use the **inferSchema=True** in **csv()** method to make Spark infer the schema of each column of the dataframe.

```python
#Create dataframe from csv file
#read contains reader functions for different types of dataset files
diabetes_df = spark.read.csv("diabetes.csv", header=True, inferSchema=True)
print(type(diabetes_df))
diabetes_df
```

```python
diabetes_df.printSchema()
```

We can obtain a list of column names for a DataFrame using the **DataFrame.columns** attribute.

```python
diabetes_df.columns
```

Instead of using **"DataFrame.show()"** we can get list of N rows using the **"DataFrame.head(N)"** method. By default head() returns only 1 row.

```python
diabetes_df.head(5)
```

We can select columns in a Spark Dataframe using the **Dataframe.select()** method - we provide a single column name or a list of column names which we want to select data from. The select() method returns a Spark DataFrame.

```python
diabetes_df.select('BMI').show(5)
diabetes_df.select(['Pregnancies', 'Glucose', 'BloodPressure']).show(5)
```

---
**NOTE :** PySpark Dataframes donot have **.loc** and **.iloc** access operations and these are only for pandas dataframes

---
Another way to select columns is by using the index operator and providing a column name or column names. If a single column name is provided, the object returned is of type **pyspark.sql.classic.column.Column** - **Column objects donot have "show()" and "head()" methods**.

If multiple columns are provided, the datatype is a DataFrame.

```python
print(type(diabetes_df['Age']), '\n')
#The following code will throw error
#diabetes_df['Age'].show()

#Selecting multiple columns
diabetes_df['Age', 'BMI'].show(5)
```

---
**TIP :** For single column selection it is preferable to use the **select()** method since using index operator returns the Column object which doesnt have a **show()** method.

---

To get the datatypes for each column of the dataframe we use **"DataFrame.dtypes"** attribute. This gives us a list of tuples - each tuple contains the column name and the datatype of the column.

```python
diabetes_df.dtypes
```

We can also use the **"DataFrame.describe()"** method to get a summary on the data inside the Dataframe. This returns a dataframe object and can be viewed using **"Dataframe.show()"** method.

describe() generates the following statistical values for summarizing the data :
* count - count of values(null included)
* mean - average value of the column
* stddev - standard deviation
* min - minimum value
* max - maximum value

```python
print(diabetes_df.describe(), type(diabetes_df.describe()), sep='\n')
print()
diabetes_df.describe().show()
```
